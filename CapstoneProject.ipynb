```python
import pandas as pd
import numpy as np
```

```python
# =========================
# Task 1: Creating DataFrames
# =========================

# Project DataFrame
project_data = {
    "ID": ["A001","A002","A003","A004","A005","A002","A005","A003","A001","A003","A001","A004","A004","A005"],
    "Project": ["Project 1","Project 2","Project 3","Project 4","Project 5","Project 6","Project 7",
                "Project 8","Project 9","Project 10","Project 11","Project 12","Project 13","Project 14"],
    "Cost": [1002000,2000000,4500000,5500000,np.nan,680000,400000,350000,np.nan,300000,2000000,1000000,3000000,200000],
    "Status": ["Finished","Ongoing","Finished","Ongoing","Finished","Failed","Finished",
               "Failed","Ongoing","Finished","Failed","Ongoing","Finished","Finished"]
}

df_project = pd.DataFrame(project_data)


# Employee DataFrame
employee_data = {
    "ID": ["A001","A002","A003","A004","A005"],
    "Name": ["John Alter","Alice Luxumberg","Tom Sabestine","Nina Adgra","Amy Johny"],
    "Gender": ["M","F","M","F","F"],
    "City": ["Paris","London","Berlin","Newyork","Madrid"],
    "Age": [25,27,29,31,30]
}

df_employee = pd.DataFrame(employee_data)


# Seniority Level DataFrame
seniority_data = {
    "ID": ["A001","A002","A003","A004","A005"],
    "Designation Level": [2,2,3,2,3]
}

df_seniority = pd.DataFrame(seniority_data)


# Display DataFrames
print("Project DataFrame")
display(df_project)

print("Employee DataFrame")
display(df_employee)

print("Seniority Level DataFrame")
display(df_seniority)
```

```python
# Save DataFrames to CSV files
df_project.to_csv("project.csv", index=False)
df_employee.to_csv("employee.csv", index=False)
df_seniority.to_csv("seniority.csv", index=False)

print("Task 1 completed and saved!")
```

```python
# ==========================================
# Task 2: Fill Missing Cost using Running Average
# ==========================================

# Variable to store sum of previous costs
total = 0

# Variable to count valid costs
count = 0

# Loop through rows
for i in range(len(df_project)):
    
    # If cost is not missing
    if not pd.isna(df_project.loc[i, "Cost"]):
        
        # Add to total
        total += df_project.loc[i, "Cost"]
        
        # Increase count
        count += 1
    
    # If cost is missing
    else:
        
        # Calculate running average
        avg = total / count
        
        # Replace missing value
        df_project.loc[i, "Cost"] = avg


# Display after filling
print("After Filling Missing Values:")
display(df_project)
```

```python
# Save updated project file
df_project.to_csv("project.csv", index=False)

print("Task 2 completed and saved!")
```

```python
# ==========================================
# Task 3: Split Name into First and Last Name
# ==========================================

# Split the Name column into two parts
df_employee[["First Name", "Last Name"]] = df_employee["Name"].str.split(" ", expand=True)

# Drop old Name column
df_employee = df_employee.drop("Name", axis=1)

# Display after splitting
print("After Splitting Name:")
display(df_employee)
```

```python
# Save updated employee file
df_employee.to_csv("employee.csv", index=False)

print("Task 3 completed and saved!")
```

```python
# ==========================================
# Task 4: Merge Project and Employee Data
# ==========================================

# Merge on ID column
df_merge1 = pd.merge(df_project, df_employee, on="ID", how="inner")

# Display result
print("After First Merge:")
display(df_merge1)
```

```python
# ==========================================
# Merge with Seniority Data
# ==========================================

# Final merge
df_final = pd.merge(df_merge1, df_seniority, on="ID", how="inner")

# Display final merged data
print("Final Merged Data:")
display(df_final)
```

```python
# Save merged data
df_final.to_csv("final_data.csv", index=False)

print("Task 4 completed and saved!")
```

```python
# =========================
# Task 5: Add Bonus Column
# =========================

# Create Bonus column
df_final["Bonus"] = 0

# Loop through rows
for i in range(len(df_final)):
    
    # If project is finished
    if df_final.loc[i, "Status"] == "Finished":
        
        # Calculate 5% bonus
        df_final.loc[i, "Bonus"] = df_final.loc[i, "Cost"] * 0.05
    
    # Otherwise bonus is 0
    else:
        df_final.loc[i, "Bonus"] = 0


# Display result
print("After Adding Bonus:")
display(df_final)
```

```python
# Save updated file
df_final.to_csv("final_data.csv", index=False)
print("Task 5 completed and saved!")
```

```python
# ==================================
# Task 6: Demote Designation for Failed
# ==================================

# Loop through rows
for i in range(len(df_final)):
    
    # If project status is Failed
    if df_final.loc[i, "Status"] == "Failed":
        
        # Increase designation level by 1 (demotion)
        df_final.loc[i, "Designation Level"] += 1


print("After Demotion:")
display(df_final) 
```

```python
# ==================================
# Delete Invalid Designations
# ==================================

# Keep only rows where designation <= 4
df_final = df_final[df_final["Designation Level"] <= 4]

print("After Deleting Designation > 4:")
display(df_final)
```

```python
# Save updated file
df_final.to_csv("final_data.csv", index=False)

print("Task 6 completed and saved!")
```

```python
# ==================================
# Task 7 : Add Mr. / Mrs. to First Name
# ==================================

# Loop through rows
for i in range(len(df_final)):
    
    # If Male
    if df_final.loc[i, "Gender"] == "M":
        
        df_final.loc[i, "First Name"] = "Mr. " + df_final.loc[i, "First Name"]
    
    # If Female
    elif df_final.loc[i, "Gender"] == "F":
        
        df_final.loc[i, "First Name"] = "Mrs. " + df_final.loc[i, "First Name"]


print("After Adding Mr/Mrs:")
display(df_final)
```

```python
# ==================================
# Drop Gender Column
# ==================================

df_final = df_final.drop("Gender", axis=1)

print("After Dropping Gender:")
display(df_final)
```

```python
# Save updated file
df_final.to_csv("final_data.csv", index=False)

print("Task 7 completed and saved!")
```

```python
# ==================================
# Task 8 : Promote Based on Age
# ==================================

# Loop through rows
for i in range(len(df_final)):
    
    # If age > 29
    if df_final.loc[i, "Age"] > 29:
        
        # Decrease designation level by 1 (promotion)
        df_final.loc[i, "Designation Level"] -= 1


print("After Promotion:")
display(df_final)
```

```python
# Save updated file
df_final.to_csv("final_data.csv", index=False)

print("Task 8 completed and saved!")
```

```python
# ==================================
# Task 9 : Calculate Total Project Cost
# ==================================

# Dictionary to store total cost
total_cost = {}

# Loop through rows
for i in range(len(df_final)):
    
    emp_id = df_final.loc[i, "ID"]
    cost = df_final.loc[i, "Cost"]
    
    # Initialize if needed
    if emp_id not in total_cost:
        total_cost[emp_id] = 0
    
    # Add cost
    total_cost[emp_id] += cost


print("Total Cost Dictionary:")
print(total_cost)
```

```python
# ==================================
# Get First Name Mapping
# ==================================

name_map = {}

for i in range(len(df_final)):
    
    emp_id = df_final.loc[i, "ID"]
    fname = df_final.loc[i, "First Name"]
    
    # Store first occurrence
    if emp_id not in name_map:
        name_map[emp_id] = fname


print("Name Mapping:")
print(name_map)
```

```python
# ==================================
# Create TotalProjCost DataFrame
# ==================================

ids = []
names = []
costs = []

for emp_id in total_cost:
    
    ids.append(emp_id)
    names.append(name_map[emp_id])
    costs.append(total_cost[emp_id])


# Create DataFrame
TotalProjCost = pd.DataFrame({
    "ID": ids,
    "First Name": names,
    "Total Cost": costs
})

display(TotalProjCost)
```

```python
# Save file
TotalProjCost.to_csv("total_project_cost.csv", index=False)

print("Task 9 completed and saved!")
```

```python
# ==================================
# Task 10 : Find Cities Containing 'o'
# ==================================

# Convert to lowercase and check 'o'
result = df_final[df_final["City"].str.lower().str.contains("o")]

print("Employees Whose City Contains 'o':")
display(result)
```
